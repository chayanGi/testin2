Attrition:- the phenomenon of the employees leaving the company.

__Possible Reasons:-__
   - Long working hours.
   - Incompetent management
   - Sickness or Death of an employee
   - Better opportunities for employees outside the organization.
   - Negative working environment.

```python
f,ax = plt.subplots(figsize=(20, 20))
sns.heatmap(df.corr(), annot=True, linewidths=.5, fmt= '.1f',ax=ax)
```

### heatmap function
 - .corr() method returns a correlation data frame containing the correlation values between the various attributes.
 - next we pass necessary parameters to the heatmap function.
     - annot: bool
         - whether to display values on top of square or not. In order to display pass it either True or df.corr()
     - linewidths: float, optional
         - width of the lines that will divide each cell
     - ax: matplotlib Axes, optional
         - Axes in which to draw the plot, otherwise use the currently-active Axes.
     - fmt: string, optional
         - String formatting code to use when adding annotations
 - returns ax: matplotlib Axes
      - Axes object with the heatmap.
